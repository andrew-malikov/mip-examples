```python
import numpy as np
import matplotlib.pyplot as plt

def show_histogram(sample: [float], bins: int, axis: [float]):
    weights=(np.ones_like(sample))/(len(sample))
    n, bins, patches = plt.hist(sample, bins, weights=weights, facecolor='g', alpha=0.75)

    plt.xlabel('Random Numbers')
    plt.ylabel('Probability')
    plt.axis(axis)
    plt.grid(True)
    
    plt.show()
```

```python
from packages.random.src.dispersion import dispersion
from packages.random.src.expected_value import expected_value

def show_expected_value(sample: [float]):
    print("Expected-Value: ", expected_value(sample))

def show_dispersion(sample: [float]):
    print("Dispersion: ", dispersion(sample))
```

## Равномерное распределение

```python
from random_examples.uniform_random import UniformRandomGenerator

uniform_generator = UniformRandomGenerator((3,9),5)

def show_uniform_example(count):
    sample = uniform_generator.generate(count)
    
    show_expected_value(sample)
    show_dispersion(sample)
    
    show_histogram(sample, 15, [3,9,0,0.3])
    
show_uniform_example(1000)
```

```python
for count in [10, 20, 50, 100, 1000]:
    print("Count: ", count)
    show_uniform_example(count)
```

## Экспоненциальное распределение

```python
from random_examples.exponential_random import ExponentialRandomGenerator

exponential_generator = ExponentialRandomGenerator(1,5)

def show_exponential_example(count):
    sample = exponential_generator.generate(count)
    show_expected_value(sample)
    show_dispersion(sample)
    
    show_histogram(sample, 20, [0,9,0,1.2])
    
show_exponential_example(1000)
```

```python
for count in [10, 20, 50, 100, 1000]:
    print("Count: ", count)
    show_exponential_example(count)
```

## Нормальное распределение

```python
from random_examples.normal_random import NormalRandomGenerator

normal_generator = NormalRandomGenerator(1,2,8)

def show_normal_example(count):
    sample = normal_generator.generate(count)
    show_expected_value(sample)
    show_dispersion(sample)
    
    show_histogram(sample, 25, [-40, -30, 0, 0.3])
    
show_normal_example(1000)
```

```python
for count in [10, 20, 50, 100, 1000]:
    print("Count: ", count)
    show_normal_example(count)
```
